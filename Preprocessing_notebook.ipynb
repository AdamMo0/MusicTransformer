####  Reading the ABC file to extract the notes

```python
USEABLE_PARAMS = [i+":" for i in "BCDFGHIKLMmNOPQRrSsTUVWwXZ"] # These are the parameters for key  

def read_abc(path):
    keys = []
    notes = []
    with open(path) as rf:
        for line in rf:
            line = line.strip()
            if line.startswith("%"): # Skip any commments
                continue
            
            if any([line.startswith(key) for key in USEABLE_KEYS]):
#                 if(line.startswith('L')):
#                     print(line) ## Checking all L in all files 
                # After checking the all midi files, they all have the length (L) : 1/8
                keys.append(line)
            else:
                notes.append(line)

    keys = " ".join(keys)
    
    notes = "".join(notes).strip()
    notes = notes.replace(" ", "")

    if notes.endswith("|"):
        notes = notes[:-1]
    # Remove unneeded character. 
    notes = notes.replace(" \ ", "") 
    notes = notes.replace("\\", "")
    notes = notes.replace("\ ", "")
    notes = notes.replace("x8|", "") # 8 because all of the midi file has a L:1/8 that means one muted bar 
    notes = notes.replace("z8|", "") # 8 because all of the midi file has a L:1/8 that means one muted bar 

    notes = notes.strip()
    notes = " ".join(notes.split(" "))
    
    if not keys or not notes:
        return None, None

    return keys, notes

def bars_similiarity(bar1, bar2):
    distances = []
    min_distance = []
    for note1 in bar1:
        for note2 in bar2:
            # Levenshtein Distance to check the similarity between the input bars and output 
            distances.append((lev.distance(note1, note2) / (len(note1) + len(note2))))
        min_distance.append(min(distances))
            
    return sum(distances) / len(distances)

def check_repeated_bars(bars):
    # If the output bars contain the same sequence of notes in each bar
    # They are removed from the dataset
    for b1 in bars:
        if(bars.count(b1)>4):
            return True
    return False
        
        
                             
# def levenshteinDistanceDP(token1, token2):
#     distances = np.zeros((len(token1) + 1, len(token2) + 1))

#     for t1 in range(len(token1) + 1):
#         distances[t1][0] = t1

#     for t2 in range(len(token2) + 1):
#         distances[0][t2] = t2
        
#     a = 0
#     b = 0
#     c = 0
    
#     for t1 in range(1, len(token1) + 1):
#         for t2 in range(1, len(token2) + 1):
#             if (token1[t1-1] == token2[t2-1]):
#                 distances[t1][t2] = distances[t1 - 1][t2 - 1]
#             else:
#                 a = distances[t1][t2 - 1]
#                 b = distances[t1 - 1][t2]
#                 c = distances[t1 - 1][t2 - 1]
                
#                 if (a <= b and a <= c):
#                     distances[t1][t2] = a + 1
#                 elif (b <= a and b <= c):
#                     distances[t1][t2] = b + 1
#                 else:
#                     distances[t1][t2] = c + 1

#     return distances[len(token1)][len(token2)]/(len(token1)+len(token2))
```

```python
# Preproccessing Steps 
    ## Clean all notes from un-needed characters in the abc notation
    ## Split the bars into 8 input bars and 8 output bars
    ## Remove any output bars that contain that has repeated notes
    ## Remove input and output bars that are similar to each other
    
def preprocessing(folder_path):
    counter = 0
    files = os.listdir(folder_path)
    for i,file in enumerate(files):
        if(file == '.DS_Store' or file[:4] !='abc_' ):
            continue
        keys, abc = read_abc(folder_path+'/'+file)
        if abc is None:
            continue
            
        abc = abc.replace(" ", "").split("|")

        
        num_bars = len(abc) // 16

        file_index = 0
        for i in range(num_bars):
            bar1 = abc[i * 8 : (i + 1) * 8]
            bar2 = abc[(i + 1) * 8 : (i + 2) * 8]

            if len(bar1) + len(bar2) != 16:
                continue

            for i in range(num_bars):
                bar1 = abc[i * 8 : (i + 1) * 8]
                bar2 = abc[(i + 1) * 8 : (i + 2) * 8]

                if len(bar1) + len(bar2) != 16:
                    continue

                if check_repeated_bars(bar2):
                    "Entered the more than 4"
                    continue # Skip 

            sim = bars_similiarity(bar1, bar2)

            if sim < 0.45:
                counter+=1
                continue            
        

            with open(folder_path + 'abcNew/'+file+'_bar_'+str(file_index)+'.txt', "w") as f:
                new_abc = keys.replace(" ", "\n") + "\n" + "|\n".join(bar1 + bar2)
                f.write(new_abc)

            file_index += 1
```

```python

for folder in FOLDER_NAMES:
    print("====================")
    print(folder)
    preprocessing(DATASET_PATH + folder + '/abc/')
    print("====================")
    

```
