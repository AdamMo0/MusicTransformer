####  Reading the ABC file to extract the notes

```python
USEABLE_PARAMS = [i+":" for i in "BCDFGHIKLMmNOPQRrSsTUVWwXZ"] # These are the parameters for key  

def read_abc(path):
    keys = []
    notes = []
    with open(path) as rf:
        for line in rf:
            line = line.strip()
            if line.startswith("%"): # Skip any commments
                continue
            
            if any([line.startswith(key) for key in USEABLE_PARAMS]):
                if(line.startswith('T')):
                    continue # skipping the title for better tokenization
                keys.append(line)
            else:
                notes.append(line)

    keys = " ".join(keys)
    
    notes = "".join(notes).strip()
    notes = notes.replace(" ", "")

    if notes.endswith("|"):
        notes = notes[:-1]
        
    # Remove unneeded character. 
    notes = notes.replace(" \ ", "") 
    notes = notes.replace("\\", "")
    notes = notes.replace("\ ", "")
    notes = notes.replace("x8|", "") # 8 because all of the midi file has a L:1/8 that means one muted bar 
    notes = notes.replace("z8|", "") # 8 because all of the midi file has a L:1/8 that means one muted bar 

    notes = notes.strip()
    notes = " ".join(notes.split(" "))
    
    if not keys or not notes:
        return None, None

    return keys, notes

def bars_similarity(bar1, bar2):
    vectorizer = CountVectorizer()
    vectorizer.fit_transform([bar1, bar2])
    vectors = vectorizer.transform([bar1, bar2])
    similarity = cosine_similarity(vectors)[0, 1]
    return similarity

def check_repeated_bars(bars):
    # If the output bars contain the same sequence of notes in each bar
    # They are removed from the dataset
    for b1 in bars:
        if(bars.count(b1)>4):
            return True
    return False
        
   
```

```python
def bars_similiarity_1(bar1, bar2):
    distances = []
    min_distance = []
    for note1 in bar1:
        for note2 in bar2:
            # Levenshtein Distance to check the similarity between the input bars and output 
            distances.append((lev.distance(note1, note2) / (len(note1) + len(note2))))
        min_distance.append(min(distances))
            
    return sum(distances) / len(distances)
bar1 = 'C'
bar2 = 'C'
bars_similiarity(bar1,bar2)
```

```python
def bars_similarity_2(bar1, bar2):
    set1 = set(bar1)
    set2 = set(bar2)
    similarity = len(set1.intersection(set2)) / len(set1.union(set2))
    return similarity

bar1 = ['C,-[G,-C,-][C-G,-C,-][^D/2-C/2G,/2-C,/2-][^D3/2-G,3/2C,3/2-]']
bar2 = ['[^A/2D/2-^A,/2-G,/2F,/2-G,,/2-][c/2-D/2C/2-^A,/2-F,/2-G,,/2-][c/2-C/2-^A']
bars_similarity(bar1,bar2)
```

```python
from difflib import SequenceMatcher

def bars_similarity_3(bar1, bar2):
    matcher = SequenceMatcher(None, bar1, bar2)
    similarity = matcher.ratio()
    return similarity

bar1 = 'C,-[G,-C,-][C-G,-C,-][^D/2-C/2G,/2-C,/2-]'
bar2 = 'c,-[f,-a,-][a-G,-C,-][^D/4-C/4X,/2-X,/2-]'
bars_similarity(bar1,bar2)
```

```python
# Preproccessing Steps 
    ## Clean all notes from un-needed characters in the abc notation
    ## Split the bars into 8 input bars and 8 output bars
    ## Remove any output bars that contain that has repeated notes
    ## Remove input and output bars that are similar to each other
    
#Preprocessing steps for maestro dataset
def preprocessing(folder_path):
    counter = 0
    files = os.listdir(folder_path)
    for i,file in enumerate(files):
        if(file == '.DS_Store' or file[:4] !='abc_' ):
            continue
        keys, abc = read_abc(folder_path+'/'+file)
        if abc is None:
            continue
            
        abc = abc.replace(" ", "").split("|")

        
        num_bars = len(abc) // 16

        file_index = 0
        for i in range(num_bars):
            bar1 = abc[i * 8 : (i + 1) * 8]
            bar2 = abc[(i + 1) * 8 : (i + 2) * 8]

            if len(bar1) + len(bar2) != 16:
                continue

            for i in range(num_bars):
                bar1 = abc[i * 8 : (i + 1) * 8]
                bar2 = abc[(i + 1) * 8 : (i + 2) * 8]

                if len(bar1) + len(bar2) != 16:
                    continue

                if check_repeated_bars(bar2):
                    "Entered the more than 4"
                    continue # Skip 

            sim = bars_similiarity(bar1, bar2)

            if sim < 0.45:
                counter+=1
                continue            
        

            with open(folder_path + 'abcNew_notitle/'+file+'_bar_'+str(file_index)+'.txt', "w") as f:
                new_abc = keys.replace(" ", "\n") + "\n" + "|\n".join(bar1 + bar2)
                f.write(new_abc)

            file_index += 1
            
```
